# Star Wars: The Force Awakens

```python
import pandas as pd
import numpy as np

star_wars = pd.read_csv('star_wars.csv', encoding ='ISO-8859-1')
star_wars.info()
```

```python
star_wars.head()
```

```python
star_wars.columns
```

## Convert columns to boolean type

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].value_counts(dropna=False)
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].value_counts(dropna=False)
```

```python
yes_no = {'Yes': True,'No': False}

cols = ['Have you seen any of the 6 films in the Star Wars franchise?',
       'Do you consider yourself to be a fan of the Star Wars film franchise?']

for col in cols:
    star_wars[col] = star_wars[col].map(yes_no)
    
    
```

```python
star_wars.head()
```

```python
star_wars = star_wars.rename(columns={"Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1"})

columns = star_wars.columns[3:9]

convert = {'Star Wars: Episode I  The Phantom Menace': True,
           'Star Wars: Episode II  Attack of the Clones': True,
           'Star Wars: Episode III  Revenge of the Sith': True,
           'Star Wars: Episode IV  A New Hope': True,
           'Star Wars: Episode V The Empire Strikes Back': True,
           'Star Wars: Episode VI Return of the Jedi': True,
          np.nan: False}

for col in columns:
    star_wars[col] = star_wars[col].map(convert)
    
```

```python
star_wars[columns].head()
```

## Rename columns

Using the function dataframe.rename() to rename all columns needed to change name, other columns will not be changed.

**Noticed:** align the changed dataframe to original dataframe

```python
new_name={'Unnamed: 4':'seen_2', 'Unnamed: 5': 'seen_3', 
          'Unnamed: 6':'seen_4', 'Unnamed: 7': 'seen_5',
          'Unnamed: 8':'seen_6'}
star_wars = star_wars.rename(columns=new_name)
```

```python
star_wars[star_wars.columns[9:15]]=star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rank_name = {'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.':'ranking_1',
             'Unnamed: 10':'ranking_2', 'Unnamed: 11':'ranking_3', 
             'Unnamed: 12':'ranking_4', 'Unnamed: 13':'ranking_5',
             'Unnamed: 14':'ranking_6'}

star_wars = star_wars.rename(columns=rank_name)
```

## Calcus the ranking scores

```python
import matplotlib.pyplot as plt

mean_score = star_wars[star_wars.columns[9:15]].mean()
mean_score
```

```python
%matplotlib inline

mean_score.plot(kind='bar',color='dodgerblue',
               title='Mean ranking score for each Episode')
```

According to the bar chart, we can find that Star Wars Episode II and III are more popular among the respondemnts, gained higher mean scores.

## Calcus how many people seen each episode

```python
seen_ = star_wars[star_wars.columns[3:9]].sum()

seen_
```

```python
plt.bar([1,2,3,4,5,6],seen_, alpha=0.5, color='g')
plt.xlabel('Star Wars Episode')
plt.ylabel('Number of seen')
plt.title('Total number of seen for each episode')
```

Upon examined the chart, it's easy to find that most people have ever seen the episode 5, then the 6. Although Episodes 2 and 3 are more popular, fewer people saw these two episodes.

## Segment data in to two groups

### 1. Split based on 'fan of star wars'

```python
sw_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']== True]
not_sw_fan = star_wars[star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?']== False]
```

```python
fig = plt.figure(figsize=(16,16))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)

x= [1,2,3,4,5,6]

data1 = sw_fan[sw_fan.columns[3:9]]
data11 = sw_fan[sw_fan.columns[9:15]]

data2 = not_sw_fan[sw_fan.columns[3:9]]
data22 = not_sw_fan[sw_fan.columns[9:15]]

ax1.bar(x,data1.sum(),color='teal')
ax1.set_title('Number of seen each episode by Fan of SW')

ax2.bar(x, data2.sum(),color='tomato')
ax2.set_title('Number of seen each episode by not Fan of SW')

ax3.bar(x,data11.mean(),color='teal')
ax3.set_title('Mean score of each episode by Fan of ST')

ax4.bar(x, data22.mean(),color='tomato')
ax4.set_title('Mean score of each episode by not Fan of ST')
```

We split the groups by if the respondents are fan of Start Wars film or not. Based on the results:

- For each episode, the number of fan seen is higher than the number of not fan seen;

- Episode 5 and 6 still have the most times of seen;

- Fans' favourite episodes are followed by 3,2,1, and the differences among the three episodes are not much.

- For not fans, their favourite episode is 3 and then 4 rather than 2. The difference are obvious.

### 2. Split based on 'fan of star trek'

```python
Yes = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == "Yes"]
No = star_wars[star_wars['Do you consider yourself to be a fan of the Star Trek franchise?'] == "No"]
```

```python
Yes1 = Yes[Yes.columns[3:9]]
Yes2 = Yes[Yes.columns[9:15]]

No1 = No[No.columns[3:9]]
No2 = No[No.columns[9:15]]

x =[1,2,3,4,5,6]

fig2 = plt.figure(figsize=(20,20))

bx1 = fig2.add_subplot(221)
bx2 = fig2.add_subplot(222)
bx3 = fig2.add_subplot(223)
bx4 = fig2.add_subplot(224)

bx1.bar(x, Yes1.sum(), color='teal')
bx1.set_title('Number of seen each episode by Fan of ST ')

bx2.bar(x, No1.sum(), color='tomato')
bx2.set_title('Number of seen each episode by not Fan of ST')

bx3.bar(x, Yes2.mean(), color='teal')
bx3.set_title('Mean score of each episode by Fan of ST')

bx4.bar(x, No2.mean(), color='tomato')
bx4.set_title('Mean score of each episode by not fan of ST')
```

In this section, we spplit the data by if the respondents are fans of Star Trek or not. The results are:

- Episodes 5 and 5 have the most seen votes by respondents either of them are fans of ST or not;

- Either fans of ST or not, their favourite episode is 3, and the following is 2.

- For fans, their degree of liking episode 2 and 3 are similar, while for not fans of st, it obviouly that they are much more like episode 3.

### 3. Split based on Gender

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
males1 = males[males.columns[3:9]]
males2 = males[males.columns[9:15]]

females1 = females[females.columns[3:9]]
females2 = females[females.columns[9:15]]

x =[1,2,3,4,5,6]

fig3 = plt.figure(figsize=(20,20))

cx1 = fig3.add_subplot(221)
cx2 = fig3.add_subplot(222)
cx3 = fig3.add_subplot(223)
cx4 = fig3.add_subplot(224)

cx1.bar(x, males1.sum(), color='teal')
cx1.set_title('Number of seen each episode by Males ')

cx2.bar(x, females1.sum(), color='tomato')
cx2.set_title('Number of seen each episode by Females')

cx3.bar(x, males2.mean(), color='teal')
cx3.set_title('Mean score of each episode by males')

cx4.bar(x, females2.mean(), color='tomato')
cx4.set_title('Mean score of each episode by not females')
```

In this section, we split the data by gender.

- The films are a little more attractive to males than to females. 
- For males, the mean scores for episodes 1, 2, and 3 are really close, meaning that males have almost same preference to them;
- For females, their favourit episode is 3, which wins a respectively higher score than other episodes.

## Segment data by other index

### 1. Education

```python
star_wars['Education'].value_counts(dropna=False)
```

```python
mean_score = star_wars.pivot_table(index='Education',
                                   values=star_wars.columns[9:15],
                                  dropna=False)
mean_score
```

```python
mean_score.plot(kind='bar',
                color=['teal','cadetblue','skyblue',
                       'lightslategrey','lightcyan','silver'],
               title='Mean score evaluated by Education',
               figsize=(12,8),rot=30)
```

When partition the data by 'Education' level, we could find that:

- For Bachlor degree, Graduate degree, and some college of Associate degress, their score evaluation are similar following the rank 3-2-1-4-6-5;

- For High School degree respondents, they are more like episode 3-1-2-6-4-5. Although this group's favourite episode is still III, but they only give it a 4.126 points, which is lower than other groups scores.

- For the group with education level Less than high school degree, they have different taste from other groups. They are much like episode I and II. Additionlly, they gave very extreme scores for different episodes: Episode II - 5.00 and Episode V - 1.00 point. 

The reason for results of the 'Less than high school degree' group may due to the characters of the young is not yet formed, or they didn't get enough educated to be objective to something.
